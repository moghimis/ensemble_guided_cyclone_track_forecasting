Make sure that the correct set of pickle data-files (containing the enslbe and best-track data) are loaded. Set the variable `ex` to choose the particular test case from list of cases available in the previously written pickle files (by the script `generate_iterative_ensemble_test_data`). 

*User configurable parameters:*

    1. ex: Test case number

    2. forecast_periods: List of times (in hours) at which the forecast is available (in steps of 6hrs for the case of GEFS ensemble forecasts)

    3. obs_times: List of observation times in hours.

    4. senFPr: Sensor footprint radius in meters
    
    5. obsErr: Error in raw-position estimation in meters-squared. 

(Algorithm parameters)

    6. r_f: Reward factor (related to penalization factor for penalizing ensemble tracks close to failed observations)

    7. extrpl_lastKnown_thresh: Threshold time-period since the last time the storm was seen in hours.  

    8. extrpl_numKnown_thresh: Threshold gap between the segments in hours.

    9. regr_seg_gap_hrs: Threshold gap between the segments in hours.

    10. regr_subseg_length_hrs: Subsegment length in hours.

    11. wetp: Weight assigned to the slope of the optimal mean ensmble track during extrapolation (1-wetp) is assigned to the slope from the last known storm-positions

```python
def lonlat_Coords(x, y, src_crs):
    """ Get the (lon,lat) coordinates given the projected coordinates."""
    LON = []
    LAT = []
    for k in range(0,len(x)):
        (_lon, _lat) = ccrs.PlateCarree().transform_point( x[k], y[k], src_crs)
        LON.append(_lon)
        LAT.append(_lat)
    return [LON, LAT]
```

### Skill versus forecast period 

```python
%matplotlib notebook
#calling it a second time may prevent some graphics errors
%matplotlib notebook  
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

plt.rcParams["font.family"] = "Times New Roman"

levels = [0, 1, 2, 3, 4, 5]
colors = ['olive', 'green', 'brown', 'yellow', 'blue']
cmap, norm = matplotlib.colors.from_levels_and_colors(levels, colors)


SMALL_SIZE = 8
MEDIUM_SIZE = 16
BIGGER_SIZE = 18

plt.rc('font', size=MEDIUM_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=MEDIUM_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=MEDIUM_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=MEDIUM_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=MEDIUM_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title


fp = [6,12,18,24,30,36,42,48]
s = []
additional_cap = []
for _fp in fp:
    x1 = sum(algo_results['d_aemn_bt'][0:len(obs_times[obs_times<=_fp])])[0]
    x2 = sum(algo_results['d_obs_bt'][0:len(obs_times[obs_times<=_fp])])[0]
    s.append(100.0 * (x1-x2)/x1)
    cap1 = sum(algo_results['d_aemn_bt'][0:len(obs_times[obs_times<=_fp])] < senFPr)[0]
    cap2 = sum(algo_results['d_obs_bt'][0:len(obs_times[obs_times<=_fp])] < senFPr)[0]
    additional_cap.append(cap2 - cap1)
    
print(*s,sep='\n')
print(*additional_cap,sep='\n')
fig = plt.figure()
plt.plot(fp, s,'bx--' )
plt.xlabel('Forecast period [hrs]')
plt.xlim((6,48))
plt.ylim((0,100))
plt.ylabel('Skill [%]')
plt.show();
```
